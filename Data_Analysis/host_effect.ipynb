```python
#ok now fix add the codes to summer_oly_medalcounts

df = pd.read_csv('../Data/summerOly_medal_counts.csv')
output_file = '../Data/summerOly_medal_counts_with_codes.csv'

data_list = []

for index, row in df.iterrows():
    if row['NOC'].strip() not in codes:
        print(row['NOC'])
        continue
    data_list.append({
        "Rank" : row['Rank'],
        "NOC" : row['NOC'],
        "Country Code" : codes[row['NOC'].strip()],
        "Gold" : row['Gold'],
        "Silver" : row['Silver'],
        "Bronze" : row['Bronze'],
        "Total" : row['Total'],
        "Year" : row['Year']
    })
    row['NOC']

output_df = pd.DataFrame(data_list)
output_df.to_csv(output_file, index = False)
```
